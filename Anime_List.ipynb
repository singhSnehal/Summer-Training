# **Load the data and infer Date datatypes**

```python
import pandas as pd

df = pd.read_csv('/content/drive/MyDrive/Summer Training/Day 2/kaggle/anime.csv', parse_dates=['start_date', 'end_date', 'created_at', 'updated_at'])
df['average_episode_duration'] = pd.to_timedelta(df['average_episode_duration'])

pd.options.display.max_columns = None
df.head(10)
```

# **Media type distribution**

```python
from plotly.offline import init_notebook_mode, iplot
import plotly.express as px

init_notebook_mode(connected=True)

fig = px.pie(df, 'media_type')
fig.update_traces(textposition='inside', textinfo='percent+label')
fig.show()
```

# **Score Distribution**

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(16, 8))

sns.histplot(df, x='mean')
```

# **Animes by year split by type**

```python
fig = px.histogram(df[df['start_date'].dt.year >= 1980], x='start_date', color='media_type')
fig.update_layout(bargap=0.1)
```
